```python
desires = ['i want to', 'i need to', 'i wish to', 'i demand to', 'i long to']
```

```python
reboot_synonyms = ['reboot', 'restart', 'reset', 'reestablish', 'reopen']
```

```python
sentences = ['pc', 'laptop', 'computer', 'my laptop', 'machine', 'my machine', 'system',
            'the system', 'kernel']
```

```python
count = 0
for x in reboot_synonyms:
    for y in sentences:
        print(x + ' ' + y + ', reboot')
        count += 1
print(count)
```

```python
count = 0
for z in desires:
    for x in reboot_synonyms:
        for y in sentences:
            print(z + ' ' + x + ' ' + y + ', reboot')
            count += 1
print(count)
```
