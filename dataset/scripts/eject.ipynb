```python
desires = ['i want to', 'i need to', 'i wish to', 'i demand to', 'i long to']
```

```python
open_synonyms = ['open', 'eject', 'disclose', 'remove', 'unlock', 'uncover']
```

```python
sentences = ['cd rom', 'cdrom', 'cd-rom', 'rom', 'my cd rom', 'my cdrom', 'the cd-rom', 'the rom']
```

```python
count = 0
for x in open_synonyms:
    for y in sentences:
        print(x + ' ' + y + ' ,eject')
        count += 1
print(count)
```

```python
count = 0
for z in desires:
    for x in open_synonyms:
        for y in sentences:
            print(z + ' ' + x + ' ' + y + ' ,eject')
            count += 1
print(count)
```
