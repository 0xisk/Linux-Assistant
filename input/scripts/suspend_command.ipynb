```python
desires = ['i want to', 'i need to', 'i wish to', 'i demand to', 'i long to']
```

```python
suspend_synonyms = ['suspend', 'sleep', 'bedtime', 'nap', 'siesta', 'snooze']
```

```python
sentences = ['pc', 'laptop', 'computer', 'my laptop', 'machine', 'my machine', 'system',
            'the system', 'kernel']
```

```python
count = 0
for x in suspend_synonyms:
    for y in sentences:
        print(x + ' ' + y + ', systemctl suspend')
        count += 1
print(count)
```

```python
count = 0
for z in desires:
    for x in suspend_synonyms:
        for y in sentences:
            print(z + ' ' + x + ' ' + y + ', systemctl suspend')
            count += 1
print(count)
```
