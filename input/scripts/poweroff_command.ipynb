```python
desires = ['i want to', 'i need to', 'i wish to', 'i demand to', 'i long to']
```

```python
turnoff_synonyms = ['turnoff', 'turn off', 'poweroff', 'power-off', 'shutdown', 'close',
                    'terminate']
```

```python
sentences = ['pc', 'personal computer', 'laptop', 'computer', 'my laptop', 'machine', 'my machine', 'system',
            'the system', 'kernel']
```

```python
count = 0
for x in turnoff_synonyms:
    for y in sentences:
        print(x + ' ' + y + ', poweroff')
        count += 1
print(count)
```

```python
count = 0
for z in desires:
    for x in turnoff_synonyms:
        for y in sentences:
            print(z + ' ' + x + ' ' + y + ', poweroff')
            count += 1
print(count)
```
